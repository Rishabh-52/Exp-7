```python
#2.Factorial
n=int(input("Enter the number whose factorial is required:"))
fact=1
while n>0:

  fact=fact*n
  n=n-1
print("factorial=",fact)
```

```python
#3(a).Fibonnacci series using while loop
n=int(input("Enter the number of terms:"))
a=0
b=1
i=1
while i <=n:
  print(a,end =" ")
  c=a+b
  a=b
  b=c
  i+=1
```

```python
#3(b).Fibonnaci series where series value shouldn't be more than 10
limit=int(input("Enter limits :"))
a = 0
b = 1
while a <=limit:
  print(a, end =" ")
  a, b = b , a+b
```

```python
#4.Reverse of a number
num=int(input("Enter a number:"))
rev = 0
while num > 0:
  digit = num %10
  rev = rev *10 + digit
  num//=10
print("Reversed number =", rev)
```

```python
#5.palindrome number
num=int(input("Enter a number:"))
temp = num
rev = 0
while num > 0 :
  rev = rev * 10 + num % 10
  num//=10
if temp==rev:
    print("palindrome")
else:
    print("Not Palindrome")
```

```python
#6.Palindrome letter
s = input("Enter a string:")
i , j = 0 , len(s)-1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i+=1
  j-=1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
#7.Check Palindrome number
st=input("enter the string :")
rev=st[::-1]
if(st==rev):
    print("palindrome")
else:
    print("not palindrome")
```

```python
#8.count digits in a number
num=int(input("Enter a number:"))
count = 0
while num > 0:
  count+=1
  num//=10
print(count)
```

```python
#9.exit the loop when i = 3
i = 1
while i < 6:
  print(i)
  if i==3:
    break
  i+=1
```

```python
#10.Search an element in a list
nums= [10 , 20 , 30 , 40 , 50]
key = int(input("Enter element to search: "))
i= 0
while i< len(nums):
  if nums[i]==key:
    print("Element found at index", i)
    break
  i +=1
else:
  print("Element not found")
```
